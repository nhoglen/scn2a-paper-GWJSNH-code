```python
import os
import numpy as np
import pandas as pd
```

```python
all_meta_files = ['metadata_intros_v4.csv','metadata_TMs.csv','metadata_reunions.csv',
                  'metadata_ppts_v2.csv','metadata_aggressions.csv']

bigtab = pd.read_csv(all_meta_files[0])
biglist = bigtab.PairTag

incompleteTags = []

for metadata_file in all_meta_files[1:]:
    meta = pd.read_csv(metadata_file)
    matchIdx = biglist.isin(meta.PairTag)
    missing = biglist[~matchIdx]
    for tag in missing:
        if tag not in incompleteTags:
            incompleteTags.append(tag)
```

```python
# --- Go through each metadata file and mark each specified tag as incomplete ---
incPd = pd.Series(incompleteTags)
bigtab[bigtab.PairTag.isin(incPd)]
```

```python
# --- Make FullTimeline columns ---

# loop over all the tables
# make a column of Trues the size of the table
# check whether PairTag is in incomplete and set those FullTimeline indicators to False

for metadata_file in all_meta_files:
    meta = pd.read_csv(metadata_file)
    nrAssays = meta.shape[0]
    ft = [True]*nrAssays
    meta.FullTimeline = ft   
    meta.loc[meta.PairTag.isin(incPd),'FullTimeline'] = False
    meta.to_csv(metadata_file,index=False)
```
