```python
import numpy as np
import pandas as pd
import os
from nwb_utils import *
from behavior_error_checks import *
from vak_utils import convert_time_with_offset
```

```python
# ----- Set up paths and files -----
cohort_tag = 'ScnVoc'

rootpath = 'M:\\scn2a-paper-GWJSNH'

nwbfile_path = os.path.join(rootpath,f'{cohort_tag}_nwb')
analysis_path = os.path.join(rootpath,f'{cohort_tag}_analysis')
acq_path = os.path.join(rootpath,f'{cohort_tag}_acquisition')
pred_path = os.path.join(rootpath,f'{cohort_tag}_predictions')
paths = [nwbfile_path,analysis_path,acq_path,pred_path]
for pathi in paths:
    if not os.path.isdir(pathi):
        os.makedirs(pathi)

metafile = 'metadata_ScnVoc_20241231.csv'
meta = pd.read_csv(os.path.join(analysis_path,metafile),sep=',')
```

```python
meta.head()
```

```python
# ----- Start times for sound files -----
audstarts = convert_time_with_offset(meta.VideoSyncPoint1,meta.VideoAssayStart,meta.AudioSyncPoint1)
meta.AudioAssayStart = audstarts
```

```python
# ----- Plotting colors -----
allcolors = []

for i,assay in enumerate(meta.PairTag):
    gt = meta.FocalGT[i]
    if gt=='WT':
        color = [0,0,0]
    else:
        color = [128, 128, 128]
    allcolors.append(color)
    
meta.FocalColor = allcolors
```
